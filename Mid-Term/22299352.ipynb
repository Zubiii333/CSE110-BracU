**Solution to the problem stated in the question paper 'B' is given below**

```python
encoded = input("Enter the encoded text: ")
decoded = ''
for i in encoded:
  ascii = ord(i)
  if ascii >= ord("a") or ascii <= ord("z"):
    new = chr(ascii - 5)
    if ord(new) < 97:
      new = chr(ascii+12)
      decoded += new
    else:
      decoded += new
    if (32 < ascii or ascii > 48 ): #Checking if there are any special characters
      new = chr(ascii - 3)
      decoded += new
    elif ascii >= ord("0") or ascii <= ord("9") : #Checking if there are any "1" in the string.
      new = chr(ascii - 3)
      decoded += new
print(decoded)
```
