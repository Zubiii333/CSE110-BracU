**Task-3:Write the Python code of a program that adds all numbers that are multiples of both 7 and 9 up to 600 (including 600) i.e. 63, 126, 189, 252, ....**

**Task-10:Write a Python program which takes a number and prints the digits from the unit place, then the tenth, then hundredth, etc**
