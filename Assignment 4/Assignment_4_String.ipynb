```python
text = str(input("Please enter a word: "))
n_text = ''
for c in text:
  ascii = ord(c)
  if ascii == ord('z'):
    n_text = n_text + 'a'
  else:
    n_text = n_text + chr(ascii+1)
print(n_text)
```

```python
text = str(input("Please enter a word: "))
temp = text[1::2]
n_text= ''
for c in temp:
  ascii = ord(c)
  n_text += chr(ascii-32)
print(n_text)
```

```python
txt = input("Please enter a word: ")
bar = ''
for chr in txt:
    if bar == '' or chr != bar[-1]:
        bar += chr
print(bar)
```

```python
#task10
txt = input("Please enter a word: ") # must include a comma(,)
text = ''
for i in range(len(txt)):
  if txt[i] == ",":
    first = txt[:i]
    text += first
    second = txt[i+2:]
    text +=second
print(text)
  
```

```python
#16
text = input("Enter a text: ")
ch = input("Enter a character: ")
new = ''
for c in text:
  if c == ch:
    continue
  else:
    new += c
print(new)
```

```python
text = input("Enter a text: ")
s = input("Enter the sign: ")
new = []
st = ''
for i in text:
  if i == s:
    new.append(st)
    st = ''
  else:
    st += i
for i in new:
  print(i)
```

```python
t = input().lower()
flag = True
t = list(t)
for i in range(len(t)):
  if t[i] >= 'a' and t[i] <= 'z' and t[i] != '':
    if flag:
      t[i] = t[i].upper()
      flag = False
    else:
      flag = True
temp = ''
for i in t:
  temp += i
print(temp)
```
